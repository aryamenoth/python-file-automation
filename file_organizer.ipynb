**Python File Automation**

A beginner Python automation project that organizes files automatically based on file type.

**Features**
- Organizes images, documents, and videos
- Uses Python file handling
- Automates repetitive file management tasks

```python
import os
import shutil

# Folder path
source_folder = "test_folder"

# File type categories
file_types = {
    "Images": [".jpg", ".png", ".jpeg"],
    "Documents": [".pdf", ".docx", ".txt"],
    "Videos": [".mp4", ".mkv"],
}

# Create folders if they don't exist
for folder in file_types.keys():
    folder_path = os.path.join(source_folder, folder)
    os.makedirs(folder_path, exist_ok=True)

# Scan files in source folder
for file_name in os.listdir(source_folder):
    file_path = os.path.join(source_folder, file_name)

    # Skip folders
    if os.path.isdir(file_path):
        continue

    # Check file type and move file
    for folder, extensions in file_types.items():
        if file_name.lower().endswith(tuple(extensions)):
            destination = os.path.join(source_folder, folder, file_name)
            shutil.move(file_path, destination)
            print(f"Moved: {file_name} → {folder}")
            break

print("Files organized successfully!")
```
